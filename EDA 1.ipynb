```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# can we combine this 2 data set
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.Country.value_counts().index
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
## pie chart   Top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation: Zomato maximum record or transaction from india after that USA and then UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation: 
1. When rating is between 4.5 to 4.9 ----> Excellent
2. When rating are between 4.0 to 4.4 ----> Very Good
3. When rating is between 3.5 to 3.9 -----> Good
4. When rating is between 3.0 to 3.4 -----> Average
5. When rating is between 2.5 to 2.9 -----> Average

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(10,10)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```
